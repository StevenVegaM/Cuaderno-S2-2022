def esPrimo(num):
    if isinstance(num, int):
        div = 2
        cont = 0
        while num > div:
            if num % div == 0:
                cont += 1
                print("False")  
            div += 1
        if cont == 0:
            print("True")
    else: 
        return "Error: no es un numero entero"

def divisores(num):
    divisor = 1
    if isinstance(num, int):
        while num >= divisor:
            if num % divisor == 0:
                print(divisor, end = ",")
            divisor += 1
    else:
        return "Error: no se ingreso un numero entero"
print(divisores(4))
